```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('honeyproduction.csv')
```

```python

df['winter_STOCK'] = df['stocks'] * 0.4
df['summer_STOCK'] = df['stocks'] * 0.6

df['winter_PVALUE'] = df['prodvalue'] * 0.4
df['summer_PVALUE'] = df['prodvalue'] * 0.6

df['winter_TP'] = df['totalprod'] * 0.4
df['summer_TP'] = df['totalprod'] * 0.6

df['consumption'] = df['totalprod']-df['stocks'] 




print(df.head(5))
```

```python
df_1 = df[df.year!=2012]
df_1
df_2 = df[df.year == 2012]
df_2
```

```python
df_3 = df_2[['stocks','prodvalue','numcol','consumption']]
df_3
```

```python
X = df[['stocks','prodvalue','numcol','consumption']].values
y = df['totalprod'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import mean_squared_error,explained_variance_score
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,predictions)   ## The red line represent the best perfect line
plt.plot(y_test,y_test,'r')
```
