```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('honeyproduction.csv')
print(df['state'].nunique())
```

```python


df['winter_STOCK'] = df['stocks'] * 0.4
df['summer_STOCK'] = df['stocks'] * 0.6

df['winter_PVALUE'] = df['prodvalue'] * 0.4
df['summer_PVALUE'] = df['prodvalue'] * 0.6

df['winter_TP'] = df['totalprod'] * 0.4
df['summer_TP'] = df['totalprod'] * 0.6

df['consumption'] = df['totalprod']-df['stocks'] 




print(df.head(5))
```

```python

df_1 = df[df.year!=2012]
df_1
df_2 = df[df.year == 2012]
df_2
```

```python
df_3 = df_2[['stocks','prodvalue','numcol','consumption']]
df_3
```

```python
X = df[['stocks','prodvalue','numcol','consumption']].values
y = df['totalprod'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
help(MinMaxScaler)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape    
```

```python
model = Sequential()
model.add(Dense(4,activation='relu'))           ## Dense layer selected 3 by x_train and activation is set to rectifier linear unit
model.add(Dense(4,activation='relu'))          ## here we selected 4 nodes in each layer and the second layer encodes the output of previous layer
                                                    ##    that are comming from the nodes and gives result
model.add(Dense(4,activation='relu'))
model.add(Dense(4,activation='relu'))                

model.add(Dense(1))
model.compile(optimizer ='adam',loss= 'mse')
```

```python
model.fit(x =X_train,y =y_train,validation_data=(X_test,y_test),batch_size=36,epochs=1700) 
```

```python
losses =pd.DataFrame (model.history.history)
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python

predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_test,predictions)   ## The red line represent the best perfect line
plt.plot(y_test,y_test,'r')
```
